```python
df.drop(["company"],axis=1,inplace=True)
df.head()
```

```python
median1 = df["agent"].median()
median1
```

```python
df["agent"]=df["agent"].replace(np.nan,median1)
```

```python
median2 = df["children"].mean()
df["children"]=df["children"].replace(np.nan,median2)
```

```python
mode1 = df["country"].mode()[0]
df["country"]=df["country"].replace(np.nan,mode1)
```

```python
df.dtypes
```

```python
df=pd.get_dummies(df,columns=["hotel","country","arrival_date_month","meal","market_segment", "distribution_channel", "reserved_room_type", "assigned_room_type", "deposit_type", "customer_type", "reservation_status", "reservation_status_date" ])
```

```python
sns.boxplot(x="adults",y="is_canceled",data=df)
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
x=df.drop(['is_canceled'],axis=1)

y=df[['is_canceled']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train, y_train)
model
```

```python
df.dtypes
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
predictions=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,predictions)
```
